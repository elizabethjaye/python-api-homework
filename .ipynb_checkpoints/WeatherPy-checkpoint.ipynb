### Analysis

* An observable trend is absolutely in relation to temperature and latitude: the closer to the equator a city is (0 degrees), the warmer the temperature is. The further north a city is, the cooler the temperature.

* Latitude and cloudiness seems to have no observable relationship - it is evenly distributed in cloudiness levels across all locations.

* Wind Speed and Latitude have a slight correlation - at very far north temperatures there is a slight uptick in higher wind speeds.


```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls

```python
#Create lists
citiesFinal = []
cloudiness = []
countries = []
date = []
humidity = []
latFinal = []
lngFinal = []
maxTemp = []
windSpeed = []

#Initialize url variable
baseurl = "http://api.openweathermap.org/data/2.5/weather?APPID="+api_key+"&units=imperial&q="

#Initialize counter
counter = 1

#Beginning of print Log
print("Beginning Data Retrieval")
print("------------------------")

#For loop to cycle through city list
for city in cities:
    url = baseurl + city
    response = requests.get(url).json()
    
    try:
        if response["message"] == "city not found":
            print("City not found. Skipping...")
    except:
        print("Processing Record #" + str(counter) + ": " + city)
        citiesFinal.append(city)
        cloudiness.append(response["clouds"]["all"])
        countries.append(response["sys"]["country"])
        date.append(response["dt"])
        humidity.append(response["main"]["humidity"])
        latFinal.append(response["coord"]["lat"])
        lngFinal.append(response["coord"]["lon"])
        maxTemp.append(response["main"]["temp_max"])
        windSpeed.append(response["wind"]["speed"])
        
        counter = counter + 1
print("------------------------")
print("Data Retrieval Complete")
print("------------------------")
```

### Convert Raw Data to DataFrame

```python
#Create DataFrame
tempDict = {"City": citiesFinal, "Cloudiness": cloudiness, "Country": countries, "Date": date, "Humidity": humidity, "Lat": latFinal, "Lng": lngFinal, "Max Temp": maxTemp, "Wind Speed": windSpeed}
weatherDF = pd.DataFrame(tempDict)

#Modify Date Column to be readable and be date only
weatherDF["Date"] = [time.strftime("%D %H:%M", time.localtime(int(row))) for row in weatherDF["Date"]]
weatherDF["Date"] = [row[:8] for row in weatherDF["Date"]]

#Export to CSV
weatherDF.to_csv(output_data_file, index=None, header=True)

#Display DataFrame
weatherDF.head()
```

### Plotting the Data

#### Latitude vs. Temperature Plot

```python
#Create Scatter
plt.scatter(weatherDF["Lat"], weatherDF["Max Temp"], c="azure", edgecolor="black")
plt.grid()
plt.xlabel('Latitude')
plt.ylabel('Max Temperature (F)')
plt.title("City Latitude vs. Max Temperature (" + str(weatherDF["Date"][1]) + ")")

#Show Image
plt.show()

#Save Image
plt.savefig("output_data/LatvsTemp.png")
```

#### Latitude vs. Humidity Plot

```python
#Create Scatter
plt.scatter(weatherDF["Lat"], weatherDF["Humidity"], c="azure", edgecolor="black")
plt.grid()
plt.xlabel('Latitude')
plt.ylabel('Humidity %')
plt.title("City Latitude vs.Humidity (" + str(weatherDF["Date"][1]) + ")")
plt.ylim(0, 110)  

#Show Image
plt.show()

#Save Image
plt.savefig("output_data/LatvsHumidity.png")
```

#### Latitude vs. Cloudiness Plot

```python
#Create Scatter
plt.scatter(weatherDF["Lat"], weatherDF["Cloudiness"], c="azure", edgecolor="black")
plt.grid()
plt.xlabel('Latitude')
plt.ylabel('Cloudiness %')
plt.title("City Latitude vs. Cloudiness (" + str(weatherDF["Date"][1]) + ")")

#Show Image
plt.show()

#Save Image
plt.savefig("output_data/LatvsCloudiness.png")
```

#### Latitude vs. Wind Speed Plot

```python
#Create Scatter
plt.scatter(weatherDF["Lat"], weatherDF["Wind Speed"], c="azure", edgecolor="black")
plt.grid()
plt.xlabel('Latitude')
plt.ylabel('Wind Speed (mph)')
plt.title("City Latitude vs. Wind Speed (" + str(weatherDF["Date"][1]) + ")")

#Show Image
plt.show()

#Save Image
plt.savefig("output_data/LatvsWindSpeed.png")
```
